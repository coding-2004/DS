```python
import pandas as pd
import numpy as np

np.random.seed(0)

df = pd.DataFrame({
    'age': np.random.randint(17, 25, 50),
    'study_hours': np.random.randint(1, 10, 50),
    'attendance': np.random.randint(50, 100, 50),
    'gpa': np.random.uniform(5, 10, 50),
    'marks': np.random.randint(40, 100, 50)
})

df.head()
```

```python
print(df.isnull().sum())
```

```python
# STEP 1: Create missing values
df.loc[0, 'marks'] = np.nan
df.loc[1, 'gpa'] = np.nan
```

```python
print(df.isnull().sum())
```

```python
# STEP 3: Handle missing values using mean
df['marks'].fillna(df['marks'].mean(), inplace=True)
df['gpa'].fillna(df['gpa'].mean(), inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df.loc[0, 'marks'] = 500
```

```python
col = 'marks'

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df[col] < lower) | (df[col] > upper)]

print("Lower:", lower)
print("Upper:", upper)
print("Outliers:\n", outliers)
```

```python
# STEP 6: handle (capping)
df[col] = np.where(df[col] > upper, upper, df[col])
df[col] = np.where(df[col] < lower, lower, df[col])

print("\nAfter capping:")
print(df.describe())
```

```python
df['study_hours_log'] = np.log1p(df['study_hours'])
```

```python
df[['study_hours', 'study_hours_log']].describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['marks']] = scaler.fit_transform(df[['marks']])
```
