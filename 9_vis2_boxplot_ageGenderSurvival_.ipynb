```python
# Import libraries

import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
```

```python
#"Titanic dataset was loaded using Seaborn built-in dataset function

df = sns.load_dataset('titanic')

# Display first 5 rows

print(df.head())
```

```python

# Dataset information

print(df.info())

# Statistical summary

print(df.describe())
```

```python
# ============================================================
# CELL 6 : CHECK MISSING VALUES
# ============================================================

print("\nMissing Values:")
print(df.isnull().sum())

# ============================================================
# CELL 7 : HANDLE MISSING VALUES IN AGE COLUMN
# ============================================================

df['age'].fillna(
    df['age'].median(),
    inplace=True
)

print("\nMissing values after filling:")
print(df['age'].isnull().sum())
```

```python
# Create box plot

plt.figure(figsize=(10,6))

sns.boxplot(
    x='sex',
    y='age',
    hue='survived',
    data=df
)

plt.title("Age Distribution by Gender and Survival")

plt.show()
```
