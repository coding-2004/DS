```python
# Import libraries

import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
```

```python
# Load iris dataset

data = load_iris()

# Create DataFrame

df = pd.DataFrame(
    data.data,
    columns=data.feature_names
)

# Add species column

df['species'] = data.target

# Display first 5 rows

print(df.head())
```

```python
# Display dataset information

print(df.info())

# Statistical summary

print(df.describe())
```

```python
# Check data types
#"Species is stored as integer labels (0,1,2), therefore pandas shows int64 data type
#but logically it represents categorical classes."
print(df.dtypes)
```

```python
# Plot histograms

df.hist(
    bins=20,
    figsize=(10,8)
)

plt.show()
```

```python
# Plot boxplots
#75% sepal length values
#  in graph 6.5 ke niche hain
plt.figure(figsize=(10,6))

sns.boxplot(data=df)

plt.title("Boxplot of Iris Features")

plt.show()
```
