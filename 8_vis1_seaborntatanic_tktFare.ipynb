```python
# Import libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load titanic dataset

df = sns.load_dataset('titanic')

# Display first 5 rows
print(df.head())
```

```python
df.info()
```

```python
# Shape of dataset

print("Shape:")
print(df.shape)

# Data types

print("\nData Types:")
print(df.dtypes)

# Missing values

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Statistical summary

print(df.describe())
```

```python
# Plot histogram of fare

plt.figure(figsize=(8,5))

plt.hist(
    df['fare'],
    bins=30
)

plt.xlabel("Fare")

plt.ylabel("Number of Passengers")

plt.title("Distribution of Titanic Fare")

plt.show()
```

```python
# Histogram using seaborn

plt.figure(figsize=(8,5))

sns.histplot(
    df['fare'],
    bins=30,
    kde=True
)

plt.title("Fare Distribution using Seaborn")

plt.show()
```

```python
# Survival count plot

sns.countplot(
    x='survived',
    data=df
)

plt.title("Survival Count")

plt.show()
```

```python
# Survival based on genderhue =
#group ke andar subgroup banana

sns.countplot(
    x='survived',
    hue='sex',
    data=df
)

plt.title("Survival Based on Gender")

plt.show()
```

```python
# Survival based on passenger class

sns.countplot(
    x='survived',
    hue='pclass',
    data=df   #Titanic dataset use करो
)

plt.title("Survival Based on Passenger Class")

plt.show()
```
